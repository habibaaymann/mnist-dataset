```python
import tensorflow as tf
from keras import Sequential
import numpy as np
mnist = tf.keras.datasets.mnist
```

```python
x = np.array([-1, 0, 1, 2, 3, 4])
y = np.array([-3, -1, 1, 3, 5, 7])
```

```python
nn=Sequential()
nn.add(tf.keras.layers.Dense(1,input_shape=[1]))
#loss fun,optimization
nn.compile(optimizer='sgd',loss='mse')
```

```python
#epochs-->number of iterations
nn.fit(x,y,epochs=450)
```

```python
print(nn.predict([10]))
```

```python
(x_train,y_train),(x_test,y_test)=mnist.load_data()
```

```python
import matplotlib.pyplot as plt
plt.imshow(x_train[400])
```

```python
#normalization
x_train,x_test=x_train/255,x_test/255
```

```python
model=Sequential()
#flatten->turn to 1D
model.add(tf.keras.layers.Flatten(input_shape=(28,28)))
model.add(tf.keras.layers.Dense(64,activation=tf.keras.activations.relu))
model.add(tf.keras.layers.Dense(64,activation=tf.keras.activations.relu))
model.add(tf.keras.layers.Dense(64,activation=tf.keras.activations.relu))
#last layer there number of classification
model.add(tf.keras.layers.Dense(10,activation=tf.keras.activations.softmax))
model.compile(optimizer='adam',loss=tf.keras.losses.sparse_categorical_crossentropy,metrics=['accuracy'])
```

```python
model.fit(x_train,y_train,epochs=3,batch_size=128)
```

```python
predictions=model(x_test)
```

```python
print(np.argmax(predictions[400]))
```
